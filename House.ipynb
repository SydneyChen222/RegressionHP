```python
import pandas as pd
from sklearn.preprocessing import label_binarize
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
trainData1 = pd.read_csv('E:/Kaggle/train.csv')
testData = pd.read_csv('E:/Kaggle/test.csv')
print(trainData1.shape)
print(testData.shape)
```

```python
trainData= trainData1.iloc[:, :-1].copy()
testData_Copy = testData.copy()
trainData.head()
```

```python
#combined_Data = pd.concat([trainData_Copy,testData_Copy], keys=[0,1])
#print(combined_Data.shape)
```

```python
total=trainData.isnull().sum().sort_values(ascending = False)
percent = (trainData.isnull().sum()/trainData.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['column', 'Percent'])
missing_data.index
missing_data.head(20)
```

```python
total=testData.isnull().sum().sort_values(ascending = False)
percent = (testData.isnull().sum()/testData.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['column', 'Percent'])
missing_data.index
missing_data.head(35)
```

```python
trainData = trainData.drop((missing_data[missing_data['Percent'] > 0.9]).index,1)
testData = testData.drop((missing_data[missing_data['Percent'] > 0.9]).index,1)
```

```python
trainData["FireplaceQu"] =trainData ["FireplaceQu"].fillna("None")
testData["FireplaceQu"] =testData ["FireplaceQu"].fillna("None")
trainData["LotFrontage"] = trainData["LotFrontage"].fillna(0)
testData["LotFrontage"] = testData["LotFrontage"].fillna(0)
```

```python
trainData["Fence"] =trainData ["Fence"].fillna("None")
testData["Fence"] =testData ["Fence"].fillna("None")
```

```python
trainData["GarageCond"] = trainData["GarageCond"].fillna("None")
trainData["GarageQual"] = trainData["GarageQual"].fillna("None")
trainData["GarageYrBlt"] = trainData["GarageYrBlt"].fillna(0)
trainData["GarageFinish"] = trainData["GarageFinish"].fillna("None")
trainData["GarageType"] = trainData["GarageType"].fillna("None")

testData["GarageCond"] = testData["GarageCond"].fillna("None")
testData["GarageQual"] = testData["GarageQual"].fillna("None")
testData["GarageYrBlt"] = testData["GarageYrBlt"].fillna(0)
testData["GarageFinish"] = testData["GarageFinish"].fillna("None")
testData["GarageType"] = testData["GarageType"].fillna("None")
```

```python
trainData["BsmtCond"] = trainData["BsmtCond"].fillna("None")
trainData["BsmtExposure"] =trainData["BsmtExposure"].fillna("None")
trainData["BsmtQual"] = trainData["BsmtQual"].fillna("None")
trainData["BsmtFinType1"] = trainData["BsmtFinType1"].fillna("None")
trainData["BsmtFinType2"] = trainData["BsmtFinType2"].fillna("None")

testData["BsmtCond"] = testData["BsmtCond"].fillna("None")
testData["BsmtExposure"] =testData["BsmtExposure"].fillna("None")
testData["BsmtQual"] = testData["BsmtQual"].fillna("None")
testData["BsmtFinType1"] = testData["BsmtFinType1"].fillna("None")
testData["BsmtFinType2"] = testData["BsmtFinType2"].fillna("None")
```

```python
trainData["MasVnrType"] = trainData["MasVnrType"].fillna("None")
trainData["MasVnrArea"] = trainData["MasVnrArea"].fillna(0)


testData["MasVnrType"] = testData["MasVnrType"].fillna("None")
testData["MasVnrArea"] = testData["MasVnrArea"].fillna(0)
```

```python
Nac=['MSZoning','Utilities','Functional', 'Exterior1st','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF', 'TotalBsmtSF', 
       'Exterior2nd','SaleType','Electrical', 'KitchenQual', 'GarageArea', 'GarageCars','BsmtHalfBath','BsmtFullBath']
```

```python
impage = SimpleImputer(missing_values=np.nan, strategy='most_frequent', fill_value=None) 
dfimpage = pd.DataFrame(impage.fit_transform(trainData[Nac]), columns=['imp_'+x for x in Nac],index=trainData.index)
trainData = pd.concat([trainData, dfimpage], axis=1)
```

```python
trainData.drop(['MSZoning','Utilities','Functional', 'Exterior1st','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF', 
       'Exterior2nd','SaleType','Electrical', 'KitchenQual', 'GarageArea', 
                   'GarageCars','BsmtHalfBath','BsmtFullBath'],axis=1,inplace=True)
trainData.head()
```

```python
categoricalFeatures=["Fence","FireplaceQu","GarageType","GarageFinish","GarageQual","GarageCond","GarageYrBlt","BsmtQual","BsmtCond","BsmtExposure","BsmtFinType1","BsmtFinType2","imp_MSZoning","MasVnrType", "Street","LotShape","LandContour","LotConfig",
                     "LandSlope","Neighborhood","Condition1","Condition2","BldgType","HouseStyle","RoofStyle",
                     "RoofMatl","imp_Exterior1st","imp_Exterior2nd","ExterQual","ExterCond","Foundation","imp_Utilities",
                     "Heating","HeatingQC","imp_Exterior1st","imp_Exterior2nd",
                     "CentralAir","imp_Electrical","imp_SaleType","imp_KitchenQual","imp_Functional","PavedDrive","SaleCondition"]
```

```python
ohe = OneHotEncoder(sparse=False, dtype=int,handle_unknown='ignore')
trainData2 = pd.DataFrame(ohe.fit_transform(trainData[categoricalFeatures]), columns=ohe.get_feature_names(),index=trainData.index)
trainData=pd.concat([trainData,trainData2],axis=1)
trainData.sample(frac=0.01)
```

```python
trainData.drop(["Fence","FireplaceQu","GarageType","GarageFinish","GarageQual","GarageCond","BsmtQual","BsmtCond","BsmtExposure","BsmtFinType1","BsmtFinType2","imp_MSZoning","MasVnrType", "Street","LotShape","LandContour","LotConfig",
                     "LandSlope","Neighborhood","Condition1","Condition2","BldgType","HouseStyle","RoofStyle",
                     "RoofMatl","imp_Exterior1st","imp_Exterior2nd","ExterQual","ExterCond","Foundation","imp_Utilities",
                     "Heating","HeatingQC","imp_Exterior1st","imp_Exterior2nd",
                     "CentralAir","imp_Electrical","imp_SaleType","imp_KitchenQual","imp_Functional","PavedDrive","SaleCondition"],axis=1,inplace=True)
trainData.head()
```

```python
Xtrain = trainData
Xtest = testData
ytrain = trainData1['SalePrice']
```

```python
from sklearn.ensemble import GradientBoostingRegressor
abc =GradientBoostingRegressor()
abc.fit(Xtrain, ytrain)
```

```python
dfimpage=pd.DataFrame(impage.transform(Xtest[Nac]),  columns=['imp_'+x for x in Nac],index=Xtest.index)
Xtest = pd.concat([Xtest, dfimpage], axis=1)
Xtest.drop(['MSZoning','Utilities','Functional', 'Exterior1st','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF', 
       'Exterior2nd','SaleType','Electrical', 'KitchenQual', 'GarageArea', 
                   'GarageCars','BsmtHalfBath','BsmtFullBath'],axis=1,inplace=True)
Xtest.head()
```

```python
dfi=pd.DataFrame(ohe.transform(Xtest[categoricalFeatures]), columns=ohe.get_feature_names(),index=Xtest.index)
Xtest = pd.concat([Xtest, dfi], axis=1)
Xtest.drop(["Fence","FireplaceQu","GarageType","GarageFinish","GarageQual","GarageCond","BsmtQual","BsmtCond","BsmtExposure","BsmtFinType1","BsmtFinType2","imp_MSZoning","MasVnrType", "Street","LotShape","LandContour","LotConfig",
                     "LandSlope","Neighborhood","Condition1","Condition2","BldgType","HouseStyle","RoofStyle",
                     "RoofMatl","imp_Exterior1st","imp_Exterior2nd","ExterQual","ExterCond","Foundation","imp_Utilities",
                     "Heating","HeatingQC","imp_Exterior1st","imp_Exterior2nd",
                     "CentralAir","imp_Electrical","imp_SaleType","imp_KitchenQual","imp_Functional","PavedDrive","SaleCondition"],axis=1,inplace=True)
Xtest.head()
```

```python
ypred=abc.predict(Xtest)
```

```python
pred2 = pd.DataFrame({'ID':testData.Id,'SalePrice':abc.predict(Xtest)})
pred2.to_csv('E:/Kaggle/GBR1.csv', index = None)
```

```python
parameters={ 'n_estimators': range(100,500,20),'min_samples_split' : range(2,100,10),'max_depth': range(2,20,2),'min_samples_leaf':range(1,20,2),'verbose':range(0,1)}
abc_random = RandomizedSearchCV(abc,parameters,n_iter=25)
abc_random.fit(Xtrain, ytrain)
grid_parm_abc=abc_random.best_params_
print(grid_parm_abc)
```
